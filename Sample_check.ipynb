```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()
df = spark.read.csv("/Volumes/workspace/default/pyspark_sample/pyspark_sample.csv",header=True, inferSchema=True )
df.show()
#type(df.select("salary").show())
df.printSchema()
df.select("Department").show()




```

```python
df.head(2)
df.dtypes
df.describe().show()
df=df.withColumn("Experience", df["Age"]-20)
df.groupBy("Department").count().show()
df.groupBy("Department").sum("Salary").show()
df.groupBy("Department").avg("Salary").show()
df.groupBy("Department").agg({"Salary":"avg"}).show()
df.groupBy("Department").agg({"Salary":"avg","Age":"avg"}).show()
df.groupBy("Department").agg({"Salary":"avg","Age":"avg"}).withColumnRenamed("avg(Salary)","Avg_Salary").withColumnRenamed("avg(Age)","Avg_Age").show()

df.show()



```

```python
df.show()
df=df.drop("work_experience")

df=df.write.mode("overwrite").parquet("/Volumes/workspace/default/pyspark_sample")


```
